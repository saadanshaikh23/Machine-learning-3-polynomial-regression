```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_csv(r"C:\Users\saads\Downloads\archive (16)\multiple_linear_regression_dataset.csv")
dataset.head(3)
```

```python
dataset.drop("age",axis = 1, inplace = True)
```

```python
dataset.head(3)
```

```python
dataset.corr()
```

```python
plt.scatter(dataset["experience"], dataset["income"])
plt.xlabel("ecperiance")
plt.ylabel("Income")
plt.show()
```

```python
x = dataset[["income"]]
y = dataset["experience"]
```

```python
dataset["income"].shape,    dataset["experience"].shape

```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree = 3)
pf.fit(x)
```

```python
x = pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train)*100
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
prd = lr.predict(x)
```

```python
plt.scatter(dataset["experience"], dataset["income"])
plt.plot(dataset["experience"],prd, c="red")
plt.xlabel("ecperiance")
plt.ylabel("Income")
plt.legend(["org", "prd"])
plt.show()
```
